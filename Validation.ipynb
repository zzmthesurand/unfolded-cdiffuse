```python
from glob import glob

test_set = glob("./CDDiffuSE-0.1/data/clean_testset_wav/*.wav", recursive=False)
print(test_set[0].split("/")[-1])
```

```python

import warnings
from scipy.io import wavfile
from pesq import pesq
from tqdm import tqdm
import pandas as pd

stats = pd.DataFrame()
speaker_id = "232"
audio_no = "033"

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for file in tqdm(test_set):
        file = file.split("/")[-1]
        rate, ref = wavfile.read(f"./CDDiffuSE-0.1/data/clean_testset_wav/{file}")
        # step 85050
        rate, deg_UF = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/same_noise_2_indiv/model85050/test/voicebank_Noisy_Test/{file}")
        rate, deg_CDFE = wavfile.read(f"./CDiffuSE-main/output/Enhanced/control/model85050/test/voicebank_Noisy_Test/{file}")
        
        # step 19575
        rate, deg_UF19575 = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/same_noise_2_indiv/model19575/test/voicebank_Noisy_Test/{file}")
        rate, deg_CDFE19575 = wavfile.read(f"./CDiffuSE-main/output/Enhanced/control/model19575/test/voicebank_Noisy_Test/{file}")
        rate, deg_UFDN19575 = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise/model19575/test/voicebank_Noisy_Test/{file}")
        
        # step 25650
        rate, deg_UFDN25650 = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise_2_indiv/model25650/test/voicebank_Noisy_Test/{file}")
        rate, deg_CDFE25650 = wavfile.read(f"./CDiffuSE-main/output/Enhanced/control/model25650/test/voicebank_Noisy_Test/{file}")
        
        # step 10800
        rate, deg_UFDN10800 = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise_2_indiv/model10800/test/voicebank_Noisy_Test/{file}")
        rate, deg_CDFE10800 = wavfile.read(f"./CDiffuSE-main/output/Enhanced/control/model10800/test/voicebank_Noisy_Test/{file}")
        rate, deg_UFDN10800_old = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise/model10800/test/voicebank_Noisy_Test/{file}")
        rate, deg_UFDN10800_max = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise_2_max/model10800/test/voicebank_Noisy_Test/{file}")
        
        # step 3375
        rate, deg_UFDN3375_old = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise/model3375/test/voicebank_Noisy_Test/{file}")
        rate, deg_UFDN3375_old_copy = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise_copy/model3375/test/voicebank_Noisy_Test/{file}")
        rate, deg_UFDN3375_new = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise_2_max_diff_compare/model3375/test/voicebank_Noisy_Test/{file}")
        rate, deg_UFDN3375_reprocess = wavfile.read(f"./CDDiffuSE-0.1/output/Enhanced/different_noise_2_max_diff_compare/model3375/test/voicebank_Noisy_Test/{file}")
        rate, deg_CDFE3375 = wavfile.read(f"./CDiffuSE-main/output/Enhanced/control/model3375/test/voicebank_Noisy_Test/{file}")

        
        band = 'wb'

        row = {
            "Audio": file,
            "UF_same_85050": [pesq(rate, ref, deg_UF, band)],
            "CDiffuSE_85050": [pesq(rate, ref, deg_CDFE, band)],
            # 19575
            "UF_diff_19575": [pesq(rate, ref, deg_UFDN19575, band)],
            "UF_same_19575": [pesq(rate, ref, deg_UF19575, band)],
            "CDiffuSE_19575": [pesq(rate, ref, deg_CDFE19575, band)],
            # 
            "UF_diff_25650": [pesq(rate, ref, deg_UFDN25650, band)],
            "CDiffuSE_25650": [pesq(rate, ref, deg_CDFE25650, band)],
            # 10800
            "UF_diff_10800": [pesq(rate, ref, deg_UFDN10800, band)],
            "UF_diff_10800_old": [pesq(rate, ref, deg_UFDN10800_old, band)],
            "UF_diff_10800_max": [pesq(rate, ref, deg_UFDN10800_max, band)],
            "UF_CDFE_10800": [pesq(rate, ref, deg_CDFE10800, band)],
            # 3375
            "UF_diff_3375_old": [pesq(rate, ref, deg_UFDN3375_old, band)],
            "UF_diff_3375_new": [pesq(rate, ref, deg_UFDN3375_new, band)],
            "UF_diff_3375_old_copy": [pesq(rate, ref, deg_UFDN3375_old_copy, band)],
            "UF_diff_3375_reprocess": [pesq(rate, ref, deg_UFDN3375_reprocess, band)],
            "UF_CDFE_3375": [pesq(rate, ref, deg_CDFE3375, band)],
        }

        stats = pd.concat((stats, pd.DataFrame(data=row)))
```
